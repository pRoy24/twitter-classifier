```python
from cassandra.cluster import Cluster
import spacy
import pandas as pd
import operator
from datetime import datetime

import os
import re
import operator
import matplotlib.pyplot as plt
import warnings
import gensim
import numpy as np
import gensim
warnings.filterwarnings('ignore')  # Let's not pay heed to them right now

from gensim.models import CoherenceModel, LdaModel, LsiModel, HdpModel
from gensim.models.wrappers import LdaMallet
from gensim.corpora import Dictionary
from pprint import pprint

%matplotlib inline

nlp = spacy.load('en_core_web_lg')
cluster = Cluster(['172.31.66.231'])
session = cluster.connect('sharelock')

```

```python
from gensim.utils import lemmatize
from nltk.corpus import stopwords
stops = set(stopwords.words('english'))  # nltk stopwords list
banned_pos = ["VERB", "ADP", "SYM"]

def process_texts(texts):
    texts = [[word for word in line if word not in stops] for line in texts]
    texts = [bigram[line] for line in texts]

    lemma_texts = []
    for line in texts:
        line_text = ' '.join(line)
        line_doc = nlp(line_text)
        bow_list = []
        for word in line_doc:
            lemma_word = ""
            if ("http" not in word.text and not word.is_stop and word.pos_ not in [banned_pos]):
                lemma_word = word.lemma_.split('/')[0]
                
            bow_list.append(lemma_word)
        lemma_texts.append(bow_list)

    return lemma_texts
```

```python
import json
import pandas as pd
import textacy
pd.set_option('display.max_rows', 2000)
    
topic_list_query = "SELECT * from sharelock.topic_list"
topic_rows = session.execute(topic_list_query)
topic_row_list = list(topic_rows)
topic_frames = pd.DataFrame(topic_row_list)  

for idx, frame in topic_frames.iterrows():
    topic = frame['topic']
    category = frame['category']
    query = "SELECT * from sharelock.active_tweets where topic='"+topic+"'order by inserted_at desc limit 4"
    rows = session.execute(query)
    ent_dict = {}    
    sorted_json = {}
    xd = json.loads(rows[0].tweet_batch)

    sorted_result = df = pd.DataFrame(data=xd) 
    train_texts = []
    for idx, frame in sorted_result.iterrows():
        train_texts.append(gensim.utils.simple_preprocess(frame['tweet_text'], deacc=True, min_len=4))

    bigram = gensim.models.Phrases(train_texts) 

    train_texts = process_texts(train_texts)
    dictionary = Dictionary(train_texts)
    corpus = [dictionary.doc2bow(text) for text in train_texts]

    ldamodel = LdaModel(corpus=corpus, num_topics=10, id2word=dictionary)
    
    vectorizer = textacy.Vectorizer(tf_type='linear', apply_idf=True, idf_type='smooth', min_df=3, max_df=0.95, max_n_terms=100000)

    doc_term_matrix = vectorizer.fit_transform(train_texts)

    model = textacy.TopicModel('nmf', n_topics=10)
    model.fit(doc_term_matrix)
    doc_topic_matrix = model.transform(doc_term_matrix)
    print(topic)
    print(vectorizer.terms_list)

 
  
```
